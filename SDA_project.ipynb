```python
class Barang:
    def __init__(self, nama, harga, stok):
        self.nama = nama
        self.harga = harga
        self.stok = stok

    def get_nama(self):
        return self.nama

    def get_harga(self):
        return self.harga

    def get_stok(self):
        return self.stok

    def set_stok(self, stok):
        self.stok = stok


class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def peek(self):
        if not self.is_empty():
            return self.items[-1]

    def size(self):
        return len(self.items)


class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)

    def size(self):
        return len(self.items)


daftar_barang = []
daftar_transaksi = []
pesanan_barang = Stack()
antrian_barang = Queue()


def tampilkan_menu():
    print("========= MENU =========")
    print("1. Tambah Barang")
    print("2. Beli Barang")
    print("3. Tampilkan Transaksi")
    print("0. Keluar")
    print("Pilih menu: ")


def get_input_pilihan():
    pilihan = int(input())
    return pilihan


def cari_barang(nama):
    for barang in daftar_barang:
        if barang.get_nama() == nama:
            return barang
    return None


def tambah_pesanan(nama):
    pesanan_barang.push(nama)


def tambah_antrian(nama):
    antrian_barang.enqueue(nama)


def proses_pesanan():
    while not pesanan_barang.is_empty():
        nama = pesanan_barang.pop()
        print(f"Pesanan {nama} sedang diproses.")


def proses_antrian():
    while not antrian_barang.is_empty():
        nama = antrian_barang.dequeue()
        print(f"Antrian {nama} sedang diproses.")


def hitung_total_barang(daftar_transaksi):
    total_barang = 0
    for barang in daftar_transaksi:
        total_barang += 1
    return total_barang


def hitung_total_harga_barang(daftar_transaksi):
    total_harga = 0
    for barang in daftar_transaksi:
        total_harga += barang.get_harga()
    return total_harga



# Inisialisasi objek Barang
barang1 = Barang("Barang A", 10000, 5)
barang2 = Barang("Barang B", 15000, 10)
barang3 = Barang("Barang C", 20000, 3)

# Menambahkan objek Barang ke daftar_barang
daftar_barang.append(barang1)
daftar_barang.append(barang2)
daftar_barang.append(barang3)

selesai = False

while not selesai:
    tampilkan_menu()
    pilihan = get_input_pilihan()

    if pilihan == 1:
        nama_barang = input("Nama barang: ")
        harga_barang = int(input("Harga barang: "))
        stok_barang = int(input("Stok barang: "))

        barang = Barang(nama_barang, harga_barang, stok_barang)
        daftar_barang.append(barang)
        print("Barang berhasil ditambahkan.")

    elif pilihan == 2:
        nama_barang = input("Nama barang yang ingin dibeli: ")

        barang = cari_barang(nama_barang)
        if barang:
            if barang.get_stok() > 0:
                barang.set_stok(barang.get_stok() - 1)
                daftar_transaksi.append(barang)
                print("Barang berhasil dibeli.")
                tambah_pesanan(nama_barang)
                tambah_antrian(nama_barang)
            else:
                print("Stok barang habis.")
        else:
            print("Barang tidak ditemukan.")

    elif pilihan == 3:
        print("Daftar Transaksi:")
        for barang in daftar_transaksi:
            print(f"Nama: {barang.get_nama()}, Harga: {barang.get_harga()}")

    elif pilihan == 0:
        selesai = True

    else:
        print("Pilihan tidak valid.")

    print()

print("Proses Pesanan:")
proses_pesanan()

print("Proses Antrian:")
proses_antrian()

total_barang_dibeli = hitung_total_barang(daftar_transaksi)
total_harga_barang = hitung_total_harga_barang(daftar_transaksi)
print("Total barang yang dibeli:", total_barang_dibeli," dengan harga:", total_harga_barang)
```
